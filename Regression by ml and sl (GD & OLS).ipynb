# Regression through Machine learning

```python
x=df.iloc[:,0:8]    #taking all rows whlie leaving last col
y=df.iloc[:,8:]     #all rows,last col
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state= 100) #splitting data into train/test portions
```

```python
y_test
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm=lm.fit(x_train,y_train)
lm
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
y_pred=lm.predict(x_test)
y_pred
```

```python
y_error = y_test - y_pred
y_error
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

## statistical learning

It is to be noted that statsmodels does not add intercept term automatically thus we need to create an intercept to our model.

```python
import statsmodels.api as sma
X_train = sma.add_constant(x_train) ## let's add an intercept (beta_0) to our model
X_test = sma.add_constant(x_test)
```

```python
X_train
```

```python
import statsmodels.api as sm
lm2 = sm.OLS(y_train,X_train).fit()
```

```python
lm2.summary()
```

```python
y_pred2 = lm2.predict(X_test) 
y_pred2
```

#### Calculate R-Squared and Adjusted R-Squared Manually on Test data

```python
import numpy as np
y_test = pd.to_numeric(y_test.CMS, errors='coerce')
RSS = np.sum((y_pred2 - y_test)**2)
y_mean = np.mean(y_test)
TSS = np.sum((y_test - y_mean)**2)
R2 = 1 - RSS/TSS
R2

n=X_test.shape[0]
p=X_test.shape[1] - 1

adj_rsquared = 1 - (1 - R2) * ((n - 1)/(n-p-1))
adj_rsquared
```
